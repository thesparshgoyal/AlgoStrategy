```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt

ticker=['IMOEX.ME']
df=yf.download(ticker,start="2020-01-01",end="2023-05-12")
df['bnh_returns'] = np.log(df['Adj Close']/df['Adj Close'].shift(1))*100
c=df['Close']
n=len(df['Open'])
k=[]
cu=[]
for i in range(n):
    if i==0:
        cu.append(0)
        a=0
        k.append(a)
    elif i<=21:
        s=0
        for j in range(i):
            s=s+c[i-1-j]
        a=s/i
        cu.append(((c[i]-c[0])/c[0])*100)
        k.append(a)
    else:
        s=0
        for j in range(21):
            s=s+c[i-1-j]
        a=s/21
        cu.append(((c[i]-c[0])/c[0])*100)
        k.append(a)
df['Returns']=cu
df['signal'] = np.where((df['Adj Close'] > k),1,0)
df['signal'] = np.where( (df['Adj Close'] < k),-1,df['signal'])
df['position'] = df['signal'].replace(to_replace=0, method='ffill')
df['position'] = df['position'].shift(1)
df['strategy_returns'] = df['bnh_returns'] * (df['position'])
print("Buy and hold returns:",df['bnh_returns'].cumsum()[-1])
print("Strategy returns:",df['strategy_returns'].cumsum()[-1])
print("Cummulative returns:",cu[-1])
df[['bnh_returns','strategy_returns']] = df[['bnh_returns','strategy_returns']].cumsum()
df[['bnh_returns','strategy_returns','Returns']].plot(grid=True, figsize=(12, 8))
```
